```python
import pandas as pd
df = pd.read_json("bp_all.json", lines=True) #lines=True
df
```

```python
df.columns
```

```python
df.sentiment.unique()
```

```python
df['sentiment'].value_counts()
#0 - negative
#1 - neutral
#2 - positive
```

```python
df = df[['date', 'sentiment']]
df
```

```python
df = df.groupby('date').mean()
df
```

```python
df.sentiment.unique()
```

```python
df['sentiment'].value_counts()
```

```python
df = df.reset_index()
df
```

```python
df.columns
```

```python
df['date'] = pd.to_datetime(df['date'])
my_range = pd.date_range(start=df.date.min(), end=df.date.max(), freq='D')
missing_dates = my_range.difference(df['date'])
df = df.append(pd.DataFrame(missing_dates, columns=['date'])).sort_values('date')
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df = df.fillna(1)
df
```

```python
df.to_csv('BP_sentiment.csv')
```
